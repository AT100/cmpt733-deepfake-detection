#### Two type of videos

1. REAL
2. FAKE

extract every 20 frames and save them to dir `fake` or `real` by its video type.

```python
train_video_path = 'deepfake-detection-challenge/train_sample_videos/'
test_video_path = 'deepfake-detection-challenge/test_videos/'

train_save_path = 'dataset/images/train/'
test_save_path = 'dataset/images/test'

count = 0
for v in os.listdir(train_video_path):
    cap = cv2.VideoCapture(os.path.join(train_video_path,v))
    i=0
    success = True
    while(success):
        success, frame = cap.read()
        if (success == True):
            #save each 20frames
            if (i % 20 == 0):
                #if video is real
                if (data[v]['label'] == 'FAKE'):
                    cv2.imwrite(train_save_path+'fake/' + str(count)+'.jpg',frame)
                #if video is false
                if (data[v]['label'] == 'REAL'):
                    cv2.imwrite(train_save_path+'real/' + str(count)+'.jpg',frame)
        i+=1
        count += 1
    print(v + ' extract finish.')
cap.release()
print('all finish')
```
